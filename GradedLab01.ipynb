The imdb.csv file contains a dataset extracted from IMDB with several attributes. For example, the title, plot, and the language of the movies. Read the imdb.csv file in a pandas dataframe and try to answer the following questions.

##### __(b) What is the min, max, and the average value of the attribute 'cumulative worldwide gross' for all movies? [1 point].__ 

```python
import pandas as pd

df = pd.read_csv('imdb.csv')
#filtering out all non movies entries
df = df[df["kind"].isin(["tv series"])==False]
df = df[df["kind"].isin(["tv mini series"])==False]
#Splitting string at 'space' to remove the ones containing a date
df[['cumulative worldwide gross','date']] = df["cumulative worldwide gross"].str.split(" ", 1, expand=True)
#Replace all non numeric characters
df['cumulative worldwide gross'] = df['cumulative worldwide gross'].str.replace(r'\D','')
#Turning the string type to int type
df['cumulative worldwide gross'] = pd.to_numeric(df['cumulative worldwide gross'])
min_value = df['cumulative worldwide gross'].min()
max_value = df['cumulative worldwide gross'].max()
avg_value = df['cumulative worldwide gross'].mean()
print(min_value)
print(max_value)
print(avg_value)

#The max value in cumulative worldwide gross is 2797800564, the minimum value is 6540000 and the average value is 390489624.68.....
```

##### __(c) What is the min, max, and the average value of the attribute 'cumulative worldwide gross' grouped by genre of the movies? [1 point].__ 

```python
import pandas as pd

df = pd.read_csv('imdb.csv')
#filtering out all non movies entries
df = df[df["kind"].isin(["tv series"])==False]
df = df[df["kind"].isin(["tv mini series"])==False]
#Splitting string at 'space' to remove the ones containing a date
df[['cumulative worldwide gross','date']] = df["cumulative worldwide gross"].str.split(" ", 1, expand=True)
#Replace all non numeric characters
df['cumulative worldwide gross'] = df['cumulative worldwide gross'].str.replace(r'\D','')
#Turning the string type to int type
df['cumulative worldwide gross'] = pd.to_numeric(df['cumulative worldwide gross'])
df_max = df.groupby(['genres'], sort=False)['cumulative worldwide gross'].max()
df_min = df.groupby(['genres'], sort=False)['cumulative worldwide gross'].min()
#Surpressing scientific notation in 'mean' and rounding
pd.options.display.float_format = '{:.2f}'.format
df_avg = df.groupby(['genres'], sort=False)['cumulative worldwide gross'].mean()

print(df_min,df_max,df_avg)
```

##### __(a)  Read the files imdb.csv and imdb2.csv in a dataframe called imdb. Add a column which is called 'plotterms' to each item. The new column should contain all terms (lower-case and cleaned from special characters) from the plot attribute which are not contained in the stopwords_english.txt file. [1 point]__ 

```python
import pandas as pd
import re

imdb = pd.concat(map(pd.read_csv, ['imdb.csv','imdb2.csv']))

stopwords = open('stopwords_english.txt', encoding="utf8")
content = stopwords.read()
wordlist = content.split('\n') #list of words that the 'plot' should not contain
stopwords.close()
pat = '|'.join(r"\b{}\b".format(x) for x in wordlist)

imdb['plotterms'] = ''
imdb['plotterms'] = imdb['plot']
imdb['plotterms'] = imdb['plotterms'].str.lower()
imdb['plotterms'] = imdb['plotterms'].str.replace(pat,'') #Replacing all words from 'plotterms' column that are in the stopwords list
imdb['plotterms'] = imdb['plotterms'].replace(r'[^\w\s]|_', '', regex=True) #All unwanted characters removed


imdb
```

##### __(b)  Create an inverted index for the terms in the 'plotterms' column. Use a datastructure of your choice for the implementation [1 point]__ 

```python
unique_words = set(' '.join([plotterms for plotterms in imdb['plotterms']]).split())
inverted_index = {}
for word in unique_words:
    inverted_index[word] = []
    index = 0
    for plotterm in imdb['plotterms']:
        if word in plotterm:
            inverted_index[word].append(index)
        index += 1
        
```

Return the merged posting list as well as the top5 items scored by the rating attribute. [3 points]__ 

```python
# merge of inverted index ['american'] and ['dream']
# skips the indexes which don't need to be compared for example
def merge_inverted_index(index1, index2):
    merged_index = []
    # if the last common index was 160, then the next comparison will start 
    # from the next higher index from the second index
    index_for_skip = 0
    for value1 in index1:
        for j in range(index_for_skip, len(index2)):
            value2 = index2[j]
            # for example: american_index = 23 and dream_index = 38: the for loop will be skipped
            if value1 < value2:
                break
            if value1 == value2:
                merged_index.append(value1)
                index_for_skip = j + 1
    return merged_index
print(inverted_index['american'])
print(inverted_index['dream'])
american_dream_index = merge_inverted_index(inverted_index['american'], inverted_index['dream'])
print(american_dream_index)
american_dream = imdb.iloc[american_dream_index].sort_values(by=('rating'), ascending = False)
american_dream.head(5)
```

##### __(e) Extend your index to be able to rank the resulting items from Task 2d by the occurrence of search terms in the plotterms. Execute the query 'american' and 'dream' (the term 'american' as well as the term 'dream' should be contained in the plot) again and print the resulting items.  [2 points]__

```python
# add ranking based on occurence in column plotterms to the inverted index
def merge_inverted_index_with_ranking(index1, index2):
    merged_index = []
    # if the last common index was 160, then the next comparison will start 
    # from the next higher index from the second index
    index_for_skip = 0
    for value1 in index1:
        for j in range(index_for_skip, len(index2)):
            value2 = index2[j]['index']
            # for example: american_index = 23 and dream_index = 38: the for loop will be skipped
            if value1['index'] < value2:
                break
            if value1['index'] == value2:
                merged_index.append({'index': value1['index'], 'ranking': value1['ranking'] + index2[j]['ranking']})
                index_for_skip = j + 1
    # sort by ranking
    merged_index = sorted(merged_index, key=lambda d: d['ranking'], reverse=True)
    return merged_index
def create_ranked_index(inverted_index):
    ranked_inverted_index = {}
    for key in inverted_index:
        ranked_inverted_index[key] = []
        for index in inverted_index[key]:
            plotterm = imdb.iloc[index]['plotterms']
            ranked_inverted_index[key].append({'index': index, 'ranking': plotterm.count(key)})
    return ranked_inverted_index
ranked_ineverted_index = create_ranked_index(inverted_index)
print(ranked_ineverted_index['american'])
print(ranked_ineverted_index['dream'])
american_dream_index = merge_inverted_index_with_ranking(ranked_ineverted_index['american'], ranked_ineverted_index['dream'])
print(american_dream_index)
american_dream = imdb.iloc[[index['index'] for index in american_dream_index]]
american_dream.head(5)
```

##### __(f) Use the TF-IDF score and the cosine similarity to execute the query 'american' and 'dream' (the term 'american' as well as the term 'dream' should be contained in the plot) against all plots in the collection. Print the results with the corresponding ranking score. Also execute the query 'american' or 'dream' (the term 'american' or the term 'dream' should be contained in the plot).  [4 points]__

```python
import math
def occurrences(word, string):
    a = string.split(" ")
    count = 0
    for i in range(0, len(a)):
        if (word == a[i]):
            count = count + 1
    return count

def wordcounter(string):
    a = string.split()
    num_of_words = len(a)
    return num_of_words
        
def countcall(word,df):
    occurrence = []
    for i in range(len(df)):
        occurrence.append(occurrences(word,df.loc[i, 'plotterms']))
    return occurrence

#Get the total number of words per row in dataframe
def get_num_of_words(df):
    wordcount = []
    for i in range(len(df)):
        wordcount.append(wordcounter(df.loc[i, 'plotterms']))
    return wordcount

#Get the term frequency
def get_tf(list_of_occurrences, list_of_wordnumber):
    tf = [i / j for i, j in zip(list_of_occurrences, list_of_wordnumber)]
    return tf

#Get the document frequency
def get_df(total_occurrence, amount_of_documents):
    df = total_occurrence / amount_of_documents
    return df

#Get inverse document frequency
def get_idf(amount_of_documents, df):
    idf = math.log(amount_of_documents / (df+1))
    return idf

occurrence = [sum(x) for x in zip(countcall('american',american_dream), countcall('dream',american_dream))]
wordcount = get_num_of_words(american_dream)
df = get_df(sum(occurrence),len(imdb.index))
idf = get_idf(len(imdb.index), df)
american_dream['tf'] = get_tf(occurrence, wordcount)
american_dream['tf-idf'] = american_dream['tf'] * idf
american_dream = american_dream.sort_values(by=('tf-idf'), ascending = False)
american_dream
```
